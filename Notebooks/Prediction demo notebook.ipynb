```python
import pandas as pd
import joblib
import numpy as np

# =========================================================
# 1. LOAD MODEL
# =========================================================
print("Loading the Bio-Computer model...")
try:
    model = joblib.load('my_best_model.pkl')
    print("âœ… Model successfully loaded! Ready to receive signals...")
except FileNotFoundError:
    print("âŒ Error: 'my_best_model.pkl' not found. Please run the Modeling notebook first.")
    exit()

# =========================================================
# 2. SIMULATE INPUT SIGNAL
# =========================================================
feature_names = [
    'Neuron_Input_AF7', 'Neuron_Input_TP9', 'Neuron_Input_AF8', 'Neuron_Input_TP10', 
    'Synapse_Activity_Alpha', 'Synapse_Activity_Beta', 'Synapse_Activity_Delta', 
    'Metabolic_Rate_1', 'Metabolic_Rate_2', 'Ion_Channel_Flux_1', 'Ion_Channel_Flux_2', 
    'Membrane_Potential', 'Total_Input_Power', 'Synapse_Ratio'
]

# Create a sample input dataframe (Simulating a high-activity state)
sample_data = {
    'Neuron_Input_AF7': 0.85,
    'Neuron_Input_TP9': 0.12,
    'Neuron_Input_AF8': 0.91,
    'Neuron_Input_TP10': 0.22,
    'Synapse_Activity_Alpha': 0.65,
    'Synapse_Activity_Beta': 0.78,
    'Synapse_Activity_Delta': 0.10,
    'Metabolic_Rate_1': 0.88,
    'Metabolic_Rate_2': 0.12,
    'Ion_Channel_Flux_1': 0.95,
    'Ion_Channel_Flux_2': 0.05,
    'Membrane_Potential': 0.45,
    'Total_Input_Power': 2.10, 
    'Synapse_Ratio': 0.83      
}

sample_input = pd.DataFrame([sample_data], columns=feature_names)

print("\n--- New Neural Signal Received ---")
print(sample_input.to_string(index=False))

# =========================================================
# 3. PREDICT
# =========================================================
prediction = model.predict(sample_input)[0]
probability = model.predict_proba(sample_input)[0]

# Get confidence score
confidence = probability[prediction] * 100

# =========================================================
# 4. REPORT
# =========================================================
print("\n" + "="*40)
print(" >>> BIO-COMPUTER STATE ANALYSIS REPORT <<<")
print("="*40)

if prediction == 1:
    state_str = "ðŸŸ¢ ACTIVE (Processing)"
    desc = "Neural computation detected. System is operational."
else:
    state_str = "ðŸ”´ INACTIVE (Resting)"
    desc = "System is in standby mode. Low neural activity."

print(f"Current State    : {state_str}")
print(f"Model Confidence : {confidence:.2f}%")
print(f"System Status    : {desc}")
print("="*40)
```
